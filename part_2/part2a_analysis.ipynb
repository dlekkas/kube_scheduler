```python
import os
import numpy as np
import pandas as pd
```

```python
RESULTS_PATH = 'results/part2a'
time_used = 'real'
```

```python
rows = ['dedup', 'blackscholes', 'ferret', 'freqmine', 'canneal', 'splash2x-fft']
cols = ['cpu', 'l1d', 'l1i', 'l2', 'llc', 'membw']
```

```python
results = pd.DataFrame(index=['ferret', 'splash2x-fft', 'freqmine', 'blackscholes', 'dedup', 'canneal'])
times_none = []
for job_dir in os.listdir(os.path.join(RESULTS_PATH, 'ibench-none')):
    job_path = os.path.join(RESULTS_PATH, 'ibench-none', job_dir)
    measurements = pd.read_csv(os.path.join(job_path, 'measurements.csv'))
    times = measurements[time_used].mean()
    times_none.append(times)
times_none = np.array(times_none)

for fif in os.listdir(RESULTS_PATH):
    if_name = fif[len('ibench-'):]
    if if_name == 'none':
        continue
    data = []
    for job_dir in os.listdir(os.path.join(RESULTS_PATH, fif)):
        job_path = os.path.join(RESULTS_PATH, fif, job_dir)
        measurements = pd.read_csv(os.path.join(job_path, 'measurements.csv'))
        times = measurements[time_used].mean()
        data.append(times)
    results[if_name] = data / times_none
```

```python
results = results.reindex(rows)[cols]
results
```
